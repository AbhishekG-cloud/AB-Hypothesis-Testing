## Problem Statement

Suppose Amazon wants to determine whether a redesigned checkout button increases the conversion rate compared with the existing design. Since user behavior naturally varies, any observed difference between the two versions may occur due to random sampling variation rather than the new design itself.

The objective of this notebook is to use statistical hypothesis testing to determine whether the observed improvement in conversion rate is statistically significant or simply the result of chance.


## Business Question
We wish to determine whether the observed difference in conversion rates between two webpage variants is statistically significant or simply the result of random sampling variability. To answer this question, we construct a hypothesis test for the difference between two population proportions.

## Z-Test Derivation

After collecting data from an A/B experiment, we observe two sample conversion rates. However, these observed conversion rates are estimates obtained from random samples and therefore vary from one experiment to another due to sampling variability.

The objective of the Two-Proportion Z-Test is to determine whether the observed difference between the two sample conversion rates is statistically significant or whether it can reasonably be explained by random sampling variation.



### Step 1: Define the Hypotheses

We begin by defining the null and alternative hypotheses.
$$
H_0 : p_A = p_B
$$

$$
H_1 : p_A \ne p_B
$$
where

- \(p_A\) is the true conversion rate of Variant A.
- \(p_B\) is the true conversion rate of Variant B.

Under the null hypothesis, both webpage variants have the same true conversion rate. Any observed difference between the sample conversion rates is assumed to arise purely due to random sampling variability.

### Step 2: Estimate the Sample Conversion Rates

The observed conversion rate for each variant is calculated as the proportion of visitors who converted.
$$
\hat{p}_A=\frac{x_A}{n_A}
$$

$$
\hat{p}_B=\frac{x_B}{n_B}
$$
where

- \(x_A\) and \(x_B\) denote the number of conversions.
- \(n_A\) and \(n_B\) denote the total number of visitors.

The observed difference between the two variants is therefore

$$
\hat{p}_B-\hat{p}_A
$$

This observed difference serves as the point estimate of the difference between the true population conversion rates.

### Step 3: Measure the Sampling Variability

The observed difference between two sample proportions changes from one experiment to another because different random samples produce different estimates.

To determine whether the observed difference is unusually large, we first measure its expected variability using the standard error.

Under the null hypothesis, both variants are assumed to have the same true conversion rate. Therefore, we estimate this common conversion rate using the pooled proportion.
$$
\hat{p}
=
\frac{x_A+x_B}
     {n_A+n_B}
$$
The pooled proportion combines the conversion data from both variants to estimate the common population conversion rate under the null hypothesis.
$$
SE
=
\sqrt{
\hat{p}(1-\hat{p})
\left(
\frac{1}{n_A}
+
\frac{1}{n_B}
\right)
}
$$
The standard error represents the expected amount of random variation in the difference between two sample conversion rates when the null hypothesis is true.

A larger standard error indicates greater sampling variability, whereas a smaller standard error suggests that the observed difference is estimated more precisely.

### Step 4: Standardize the Difference

The observed difference alone is not sufficient to determine statistical significance.

Instead, we compare the observed difference with its expected sampling variability by dividing the difference by its standard error.

This standardized quantity is known as the Z-statistic.
$$
Z
=
\frac{
\hat{p}_B-\hat{p}_A
}{
SE
}
$$
Under the null hypothesis and for sufficiently large sample sizes, the Z-statistic approximately follows the standard normal distribution.

This property allows us to calculate a p-value and determine whether the observed difference is statistically significant.

## Visualization 1 — Sampling Distribution under H₀
### Objective

Create a Plotly visualization of the Standard Normal Distribution. This graph is the foundation for all later visualizations (rejection region, Z-statistic marker, p-value).

```python
import numpy as np
import plotly.graph_objects as go
from scipy.stats import norm
```

```python
x = np.linspace(-4, 4, 500)
y = norm.pdf(x)
fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x=x,
        y=y,
        mode="lines",
        name="Standard Normal Distribution"
    )
)
fig.update_layout(
    title="Sampling Distribution of the Z-Statistic Under H₀",
    xaxis_title="Z-Statistic",
    yaxis_title="Probability Density",
    template="plotly_white"
)
fig.add_vline(
    x=0,
    line_dash="dash",
    line_color="black"
)
```

### Interpretation

This figure represents the sampling distribution of the Z-statistic under the null hypothesis.

If the null hypothesis is true, most observed Z-statistics are expected to lie near zero due to random sampling variability. As the Z-statistic moves farther away from zero, the observed result becomes increasingly unlikely under the null hypothesis.

## Visualization 2 — Rejection Region
### Objective

Highlight the region where we reject the null hypothesis

```python
x = np.linspace(-4, 4, 500)
y = norm.pdf(x)
fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x=x,
        y=y,
        mode="lines",
        name="Standard Normal Distribution"
    )
)
fig.update_layout(
    title="Sampling Distribution of the Z-Statistic Under H₀",
    xaxis_title="Z-Statistic",
    yaxis_title="Probability Density",
    template="plotly_white"
)
fig.add_vline(
    x=0,
    line_dash="dash",
    line_color="black"
)
fig.add_vline(
    x=-1.96,
    line_dash="dash",
    line_color="red"
)

fig.add_vline(
    x=1.96,
    line_dash="dash",
    line_color="red"
)
```

```python
x_right = np.linspace(1.96, 4, 200)
y_right = norm.pdf(x_right)
fig.add_trace(
    go.Scatter(
        x=x_right,
        y=y_right,
        fill="tozeroy",
        mode="lines",
        line=dict(width=0),
        showlegend=False
    )
)
```

```python
x_left = np.linspace(-4, -1.96, 200)
y_left = norm.pdf(x_left)
fig.add_trace(
    go.Scatter(
        x=x_left,
        y=y_left,
        fill="tozeroy",
        mode="lines",
        line=dict(width=0),
        showlegend=False,
        
    )
)
fig.update_layout(
    title="Sampling Distribution with Rejection Regions"
)
```

### Business Interpretation

The shaded regions represent outcomes that are considered sufficiently unlikely if the null hypothesis is true.

If the observed Z-statistic falls inside either shaded region, the evidence against the null hypothesis is strong enough to reject it at the 5% significance level.

In an A/B test, this means the observed difference between the two variants is unlikely to be explained by random sampling variation alone.

## Visualization 3 — Observed Z-Statistic
### Objective

Mark the Z-statistic computed from your experiment on the sampling distribution.

This answers the question:

"Where does my experiment lie on the null distribution?"

```python
z_stat = 2.35
fig.add_vline(
    x=z_stat,
    line_color="green",
    line_width=3
)
fig.add_annotation(
    x=z_stat,
    y=0.25,
    text=f"Observed Z = {z_stat:.2f}",
    showarrow=True,
    arrowhead=2
)
```

### Business Interpretation

The green vertical line represents the observed Z-statistic calculated from the A/B experiment.

If this value falls within the rejection region, the observed difference between the two variants is considered statistically significant at the chosen significance level.

This visualization helps decision-makers understand not only the outcome of the hypothesis test but also how far the observed result lies from what would typically be expected under the null hypothesis.

## Visualization 4 — Confidence Interval
### Objective

Visualize the estimated difference in conversion rates along with its 95% confidence interval.

```python
fig_ci.add_vline(
    x=0,
    line_dash="dash",
    line_color="red"
)
fig_ci.update_layout(
    title="95% Confidence Interval for the Difference in Conversion Rates",
    xaxis_title="Difference in Conversion Rate",
    yaxis=dict(showticklabels=False),
    template="plotly_white"
)
```

### Business Interpretation

The point represents the estimated improvement in conversion rate, while the horizontal line represents the 95% confidence interval.

If the interval does not include zero, the observed improvement is statistically significant at the 5% significance level.

The width of the interval reflects the uncertainty in our estimate. Narrow intervals indicate more precise estimates, whereas wider intervals indicate greater uncertainty.

## Visualization 5 — Power Curve
### Objective

Show how increasing the sample size increases the probability of detecting a true effect.
Power is defined as

$$
\text{Power} = P(\text{Reject } H_0 \mid H_1 \text{ is true})
$$

In simple words,

If there really is an improvement, what is the probability that our experiment correctly detects it?**

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent
sys.path.append(str(PROJECT_ROOT))
```

```python
from src.frequentist_tests import minimum_sample_size
baseline_rate = 0.10
mde = 0.02
alpha = 0.05
powers = np.linspace(0.50, 0.99, 20)
sample_sizes = [
    minimum_sample_size(
        baseline_rate=baseline_rate,
        mde=mde,
        alpha=alpha,
        power=p
    )
    for p in powers
]
```

```python
fig_power = go.Figure()
fig_power.add_trace(
    go.Scatter(
        x=sample_sizes,
        y=powers,
        mode="lines+markers",
        name="Power Curve"
    )
)
fig_power.update_layout(
    title="Effect of Sample Size on Statistical Power",
    xaxis_title="Sample Size per Variant",
    yaxis_title="Statistical Power",
    template="plotly_white"
)
```

Organizations therefore perform sample size calculations before launching A/B tests to ensure that meaningful improvements are unlikely to be missed.
