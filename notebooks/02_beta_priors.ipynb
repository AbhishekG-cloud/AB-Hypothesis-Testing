# Introduction

In Bayesian statistics, the **Beta distribution** is commonly used to model uncertainty about an unknown probability. In the context of A/B testing and conversion rate optimization, it represents our belief about the **true conversion probability** of a variant—not the number of conversions themselves. As new data is observed, this belief is updated, allowing us to estimate the conversion rate in a principled and interpretable way.

The Beta distribution is defined by two parameters, **α (alpha)** and **β (beta)**, which determine the shape of the distribution and encode our prior belief. Different values of these parameters correspond to different levels of confidence and different expectations about the underlying conversion rate.

In this notebook, we will explore how varying α and β changes the Beta distribution, compute important summary statistics such as the **mean**, **mode**, **variance**, and **credible intervals**, construct informative priors using historical conversion data, and visualize different prior distributions to understand their practical interpretation in business decision-making.

# 3. Exploring the Non-Informative Prior: Beta(1,1)

We begin with the **Beta(1,1)** distribution, often referred to as a **uniform (non-informative) prior**. It assigns equal probability density to every possible conversion rate between 0 and 1, indicating that, before observing any data, we have no preference for one conversion rate over another.

Although commonly used when little or no prior information is available, Beta(1,1) should not be interpreted as stating that every conversion rate is equally likely in reality. Rather, it reflects that **our current knowledge provides no reason to favor one value over another**.

In this section, we will:

- Visualize the Beta(1,1) distribution.
- Compute its mean, variance, and 95% credible interval.
- Observe that the mode is undefined using the standard formula because α = β = 1.
- Interpret what this prior represents from a business perspective.

```python
alpha = 1
beta = 1

print(f"Prior: Beta({alpha}, {beta})")

mean = beta_mean(alpha, beta)
variance = beta_variance(alpha, beta)
ci = beta_ci(alpha, beta)

print(f"Mean               : {mean:.4f}")
print(f"Variance           : {variance:.4f}")
print(f"95% Credible Interval: ({ci[0]:.4f}, {ci[1]:.4f})")

try:
    mode = beta_mode(alpha, beta)
    print(f"Mode               : {mode:.4f}")
except ValueError as e:
    print(f"Mode               : Undefined ({e})")

fig = plot_beta(
    alpha=alpha,
    beta=beta,   # Use your actual parameter name here
    label="Beta(1,1): Non-Informative Prior"
)

fig.show()
```

### Business Interpretation

The **Beta(1,1)** prior represents a situation where no reliable historical information is available. Every conversion rate between 0 and 1 is considered equally plausible before observing any data.

This prior is commonly used as a neutral starting point for new products or experiments with no relevant historical evidence. As new observations become available, this prior will be updated to form the posterior distribution.
