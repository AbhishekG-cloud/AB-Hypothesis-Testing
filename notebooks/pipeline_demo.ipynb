# Bayesian A/B Testing Framework — End-to-End Demonstration

## Objective

This notebook demonstrates the complete Bayesian A/B testing pipeline developed throughout the project.

The framework integrates:

- Frequentist hypothesis testing
- Bayesian posterior inference
- Monte Carlo simulation
- Business decision recommendation

The goal is to analyze an A/B experiment from raw conversion counts to a final business recommendation using a unified statistical framework.

```python
import sys
from pathlib import Path

project_root = Path.cwd().parent
sys.path.append(str(project_root))
from src.ab_framework import ABTestFramework
from src.comparison import comparison_plot
```

```python
# Experiment Data

visitors_A = 1000
conversions_A = 52

visitors_B = 1000
conversions_B = 61

prior_alpha = 1
prior_beta = 1
```

```python
framework = ABTestFramework()
results = framework.run_full_analysis(
    n_A=visitors_A,
    conv_A=conversions_A,
    n_B=visitors_B,
    conv_B=conversions_B,
    prior_A=(1,1),
    prior_B=(1,1)
)
```

```python
frequentist = results["frequentist"]
z = results["frequentist"]["z_statistic"]
p = results["frequentist"]["p_value"]
ci = results["frequentist"]["confidence_interval"]
print("Frequentist Analysis")
print("-" * 40)
print(f"Z-statistic          : {frequentist['z_statistic']:.4f}")
print(f"P-value              : {frequentist['p_value']:.4f}")
print(f"95% Confidence Interval : {frequentist['confidence_interval']}")
```

```python
bayesian = results["bayesian"]
print("Bayesian Analysis")
print("-"*40)


print(f"Posterior Mean A : {bayesian['A']['mean']:.4f}")
print(f"Posterior Mean B : {bayesian['B']['mean']:.4f}")

print(f"Credible Interval A : {bayesian['A']['credible_interval']}")
print(f"Credible Interval B : {bayesian['B']['credible_interval']}")
```

```python
monte_carlo = results["monte_carlo"]
print("Monte Carlo Analysis")
print("-" * 40)

print(f"P(B > A)        : {monte_carlo['probability_B_beats_A']:.4%}")
print(f"Expected Uplift : {monte_carlo['expected_uplift']:.4%}")
print(f"Expected Loss   : {monte_carlo['expected_loss']:.4%}")
```

Variant B shows a higher observed conversion rate than Variant A (6.1% vs. 5.2%), but the evidence is not yet strong enough to recommend deployment. The frequentist analysis does not find a statistically significant difference (p = 0.383), and the Bayesian analysis estimates an 80.9% probability that B is better—moderate, but below common decision thresholds such as 95% or 99%. While the expected uplift is encouraging (0.9%), additional data should be collected before making a production decision.
